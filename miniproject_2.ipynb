```python
#создадим функцию для перевода строк в snake case
def snake_case(name):
    new_name = name.replace(' ', '_').lower()
    return new_name

#переименуем названия столбцов в snake case
bookings = bookings.rename(columns=snake_case)
```

```python
#Создайте колонку total_kids, объединив столбцы children и babies. 
#Для отелей какого типа среднее значение переменной оказалось наибольшим?

bookings['total_kids'] = bookings.children + bookings.babies
```

```python
#Сгруппируем данные по типу отеля и посчитаем среднее число детей у клиентов в каждой группе. 
#А затем округлим значения до 2-ух знаков и возьмем максимум
bookings \
    .groupby('hotel') \
    .agg({'total_kids': 'mean'}).round(2).max
```

```python
bookings.head()
```

```python
#Выясним среди какой группы выше Churn Rate: с детьми или без детей

bookings['has_kids'] = (bookings.total_kids != 0)
```

```python
#Выясним общее количество посетителей без детей
without_kids = bookings \
    .query('has_kids == False')
without_kids_sum = len(without_kids.index)
```

```python
#Выясним общее количество посетителей с детьми
with_kids = bookings \
    .query('has_kids == True')
with_kids_sum = len(with_kids.index)
```

```python
#Сделаем проверку. Сравним общее количество бронирований с суммой бронирований с детьми и без детей.

len(bookings.index) == (len(without_kids.index) + len(with_kids.index))
```

```python
#Выясним количество посетителей без детей, отменивших бронирование
without_kids_canceled = bookings \
    .query('has_kids == False') \
    .query('is_canceled == 1')
without_kids_canceled_sum = len(without_kids_canceled.index)
```

```python
#Выясним количество посетителей с детьми, отменивших бронирование
with_kids_canceled = bookings \
    .query('has_kids == True') \
    .query('is_canceled == 1')
with_kids_canceled_sum = len(with_kids_canceled.index)
```

```python
churn_rate_without_kids = (without_kids_canceled_sum / without_kids_sum)
churn_rate_with_kids = (with_kids_canceled_sum / with_kids_sum)
```
